# Scatter plot 
### In scatter plot the value of 2 varialbes are plotted as points on a 2-dimensional grind . Additionally you can use a third variable also

```python
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df
```

```python
plt.plot(flowers_df.sepal_length,flowers_df.sepal_width);
```

```python
sns.scatterplot([flowers_df.sepal_length],[flowers_df.sepal_width]);
```

```python
sns.scatterplot(flowers_df.sepal_length,flowers_df.sepal_width,hue= flowers_df.species,s=100)
```

## Plotting using pandas

```python
plt.title('Sepal Dimensions')
sns.scatterplot('sepal_length',
                'sepal_width',
                hue = 'species',
                s=100,
                data=flowers_df);
```

 # Heatmaps
 

```python
flights_df = sns.load_dataset("flights")
```

```python
flights_df
```

```python
flights_df = flights_df.pivot(index="month",columns="year",values="passengers")
```

```python
flights_df
```

```python
plt.title("No of passengers (1000s)")
sns.heatmap(flights_df)
```

```python
plt.title("No of passengers (1000s)")
sns.heatmap(flights_df,fmt="d",annot=True,cmap='Blues')
#annot = show the actual value fo each block
#cmap = to change color
```
